```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
dane = np.loadtxt("zadanie2.dat")
p = np.arange(0.4, 0.8, 0.01)

plt.figure(figsize = (4,4))
plt.plot(p, dane[0], label = 'L = 16', color='k')
plt.plot(p, dane[1], label = 'L = 32', color='r')
plt.xlabel('prawdopodobienstwo obsadzenia wezla')
plt.ylabel('W')
plt.grid(True)
plt.xlim(min(p), max(p))
plt.legend(frameon = False)
plt.savefig('wykres.pdf', format = 'pdf')
plt.tight_layout()
plt.show()
```
