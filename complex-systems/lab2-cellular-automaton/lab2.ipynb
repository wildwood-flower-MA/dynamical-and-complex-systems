```python
import numpy as np
import os
import matplotlib.pyplot as plt
import matplotlib.image as mgimg
import matplotlib.animation as animation
plt.rc('xtick', labelsize=20) 
plt.rc('ytick', labelsize=20)
```

```python
próba = 3

plansze = []
plansza_raw = np.loadtxt("lab2_plansze.dat")
size = np.shape(plansza_raw)
zp_raw = np.loadtxt("lab2_zapelnienie.dat")

for n in range(int(size[0]/size[1])):
    plansze.append(plansza_raw[size[1]*n:size[1]*(n+1), :])

for idx, plansza in enumerate(plansze):
    fig, ax = plt.subplots()
    ax.imshow(plansza, cmap="binary")
    ax.set_xticks(np.arange(size[1]+1) - 0.5, minor=True)
    ax.set_yticks(np.arange(size[1]+1) - 0.5, minor=True)
    ax.set_xticks([]); ax.set_yticks([])
    plt.title("B4678/S35678"+f" [iteracji: {int(zp_raw[idx][0])}, zapelnienie: {100*zp_raw[idx][1]:.2f}%]")
    plt.savefig(f"próba_{próba}_w_iteracji_{int(zp_raw[idx][0])}.png")
    plt.show()
```
